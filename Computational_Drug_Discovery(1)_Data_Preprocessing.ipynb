## Handling missing data

```python
df2 = df[df.standard_value.notna()]
df2
```

## Data pre-processing of the bioactivity data

The bioactivity data is in the IC50 unit. 
- Compounds having values of less than 1000 nM will be considered to be **active**.

- As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**.

- Compounds having values greater than 10,000 nM will be considered to be **inactive**. 

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

## Iterate the molecule_chembl_id to a list

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

## Iterate canonical_smiles to a list

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

## Iterate standard_value to a list

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

## Combine the 4 lists into a dataframe

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df4 = pd.concat([df3, bioactivity_class], axis=1)
df4
```
